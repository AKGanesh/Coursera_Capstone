```python
def get_url(lat, lng):
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
        CLIENT_ID, 
        CLIENT_SECRET, 
        VERSION, 
        lat, 
        lng, 
        radius, 
        LIMIT)
    return url



def get_results():
    results = []
    for index, row in df_clean.iterrows():
        url = get_url(row['Latitude'], row['Longitude'])
        results.append(requests.get(url).json())
    return results


results = get_results()
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
from pandas.io.json import json_normalize

def get_nearby_venues(result):
    try:
        venues = result['response']['groups'][0]['items']

        nearby_venues = json_normalize(venues) # flatten JSON

        # filter columns
        filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
        nearby_venues =nearby_venues.loc[:, filtered_columns]

        # filter the category for each row
        nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

        # clean columns
        nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

        return nearby_venues
    except:
        return None
```

```python
nearby_venues = []
for i in range(70):
    result = results[i]
    venues = get_nearby_venues(result)
    if(venues is not None):
        nearby_venues.append(venues)
#         print(len(venues))

#nearby_venues
```

```python
list_manitoba_venues = pd.concat(nearby_venues, ignore_index=True)
list_manitoba_venues = list_manitoba_venues.drop_duplicates()
list_manitoba_venues.head()
```

```python
all_manitoba_venues_unique = list_manitoba_venues['categories'].value_counts().to_frame(name='Count')
```

```python
all_manitoba_venues_unique.to_csv('all_manitoba_venues_categories.csv')
all_manitoba_venues_unique = pd.read_csv('all_manitoba_venues_categories_with_weight.csv')
list_manitoba_venues.head()
```

```python
all_manitoba_venues_unique = pd.read_csv('all_manitoba_venues_categories_with_weight.csv')
all_manitoba_venues_unique = all_manitoba_venues_unique[['Category', 'Count', 'Attraction', 'Competition']]
all_manitoba_venues_unique.head()

weights = []

for index, row in list_manitoba_venues.iterrows():
#     cat_weight = all_manitoba_venues_unique['Category'] == row['categories']
    
    cat_weight = all_manitoba_venues_unique[all_manitoba_venues_unique['Category'] == row['categories']]
    cat_weight = cat_weight[['Attraction', 'Competition']].values
    
#     print(index)
#     print(row['categories'])
    
    if len(cat_weight) == 0:
        attr = 8
        comp = 3
        weight_ = attr - comp
    else:
        attr = cat_weight[0][0]
        comp = cat_weight[0][1]
        weight_ = attr - comp

    cat_weight = [attr, comp, weight_]
    weights.append(cat_weight)

weights = pd.DataFrame(weights, columns = ['Attraction', 'Competition', 'Weight'])

list_manitoba_venues.index = weights.index
list_manitoba_venues['Attraction'] = weights['Attraction']
list_manitoba_venues['Competition'] = weights['Competition']
list_manitoba_venues['Weight'] = weights['Weight']

# df.head()

list_manitoba_venues.head()
```

# For clustering, We can make use of K-Means Clustering algorithm

```python
from sklearn.cluster import KMeans

# set number of clusters
kclusters = 20

grouped_clustering = list_manitoba_venues[['lat', 'lng']]

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(grouped_clustering)

# check cluster labels generated for each row in the dataframe
# kmeans.labels_[0:10]
kmeans.labels_
#list_manitoba_venues = list_manitoba_venues.drop(columns=['Cluster Labels'], axis=1)
list_manitoba_venues.insert(0, 'Cluster Labels', kmeans.labels_)
list_manitoba_venues.head()
```

```python
list_manitoba_venues[list_manitoba_venues['Cluster Labels'] == 1]
```

```python
kcluster_sum = list_manitoba_venues.groupby('Cluster Labels').aggregate({'Weight': 'sum', 'Competition': 'count'})
kcluster_sum = kcluster_sum.rename(index=str, columns = {'Competition':'Count'})
# print(df.shape)
kcluster_sum['avg'] =  kcluster_sum['Weight'] / kcluster_sum['Count']
kcluster_sum = kcluster_sum.sort_values(by = ['avg'], ascending = [False])
kcluster_sum.head()
```

# The cluster with the highest average value is the best location for our restaurant.

```python
best_cluster = list_manitoba_venues[list_manitoba_venues['Cluster Labels'] == 14]
print(best_cluster['lat'].mean())
best_cluster.head()
```

```python
import folium
center_of_cluster = [best_cluster['lat'].mean(), best_cluster['lng'].mean()]
map = folium.Map(location=center_of_cluster, zoom_start=12)

# for i in range(0,len(best_cluster)):
#     folium.Marker([best_cluster.iloc[i]['lat'], best_cluster.iloc[i]['lng']], popup=best_cluster.iloc[i]['name']).add_to(map)
    
# map

# create map of New York using latitude and longitude values
# map_newyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, name, categories in zip(best_cluster['lat'], best_cluster['lng'], best_cluster['name'], best_cluster['categories']):
    label = '{}, {}'.format(name, categories)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map)  
    
map
```
